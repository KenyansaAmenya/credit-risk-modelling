**Categorical Features Treatment**

```python
cr_data_categorical_treated = cr_data_copy.copy()
```

```python
# check categories in this data
cr_data_categorical_treated.groupby('person_home_ownership').count()['person_age']
```

```python
person_home_ownership = pd.get_dummies(cr_data_categorical_treated['person_home_ownership'], drop_first=True).astype(int)
person_home_ownership.head()
```

```python
loan_intent = pd.get_dummies(cr_data_categorical_treated['loan_intent'], drop_first=True).astype(int)
loan_intent.head()
```

```python
cr_data_categorical_treated['cb_person_default_on_file_binary'] = np.where(cr_data_categorical_treated['cb_person_default_on_file']=='Y',1,0)
cr_data_categorical_treated.head()
```

```python
# Remove columns which are not required
data_to_scale = cr_data_categorical_treated.drop(['person_home_ownership', 'loan_intent', 'loan_status', 'cb_person_default_on_file_binary', 'cb_person_default_on_file'], axis=1)
```

```python
data_to_scale.head()
```

**Scaling our Data**

```python
scaler = StandardScaler()
```

```python
data_to_scale.columns
```

```python
scaled_data = scaler.fit_transform(data_to_scale)
scaled_df = pd.DataFrame(scaled_data, columns=['person_age', 'person_income', 'person_emp_length', 'loan_amnt',
       'loan_int_rate', 'loan_percent_income', 'cb_person_cred_hist_length'])
scaled_df.head()
```

```python
scaled_df.shape
```

```python
cr_data_categorical_treated.head()
```

```python
scaled_data_combined = pd.concat([scaled_df,person_home_ownership,loan_intent],axis=1)
scaled_data_combined['cb_person_default_on_file'] = cr_data_categorical_treated['cb_person_default_on_file_binary']
scaled_data_combined['loan_status'] = cr_data_categorical_treated['loan_status']
scaled_data_combined.head()
```

```python
scaled_data_combined.groupby('loan_status').count()['EDUCATION']
```
