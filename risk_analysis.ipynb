**Categorical Features Treatment**

```python
cr_data_categorical_treated = cr_data_copy.copy()
```

```python
# check categories in this data
cr_data_categorical_treated.groupby('person_home_ownership').count()['person_age']
```

```python
person_home_ownership = pd.get_dummies(cr_data_categorical_treated['person_home_ownership'], drop_first=True).astype(int)
person_home_ownership.head()
```

```python
loan_intent = pd.get_dummies(cr_data_categorical_treated['loan_intent'], drop_first=True).astype(int)
loan_intent.head()
```

```python
cr_data_categorical_treated['cb_person_default_on_file_binary'] = np.where(cr_data_categorical_treated['cb_person_default_on_file']=='Y',1,0)
cr_data_categorical_treated.head()
```

```python
# Remove columns which are not required
data_to_scale = cr_data_categorical_treated.drop(['person_home_ownership', 'loan_intent', 'loan_status', 'cb_person_default_on_file_binary', 'cb_person_default_on_file'], axis=1)
```

```python
data_to_scale.head()
```

**Scaling our Data**

```python
scaler = StandardScaler()
```

```python
data_to_scale.columns
```

```python
scaled_data = scaler.fit_transform(data_to_scale)
scaled_df = pd.DataFrame(scaled_data, columns=['person_age', 'person_income', 'person_emp_length', 'loan_amnt',
       'loan_int_rate', 'loan_percent_income', 'cb_person_cred_hist_length'])
scaled_df.head()
```

```python
scaled_df.shape
```

```python
cr_data_categorical_treated.head()
```

```python
scaled_data_combined = pd.concat([scaled_df,person_home_ownership,loan_intent],axis=1)
scaled_data_combined['cb_person_default_on_file'] = cr_data_categorical_treated['cb_person_default_on_file_binary']
scaled_data_combined['loan_status'] = cr_data_categorical_treated['loan_status']
scaled_data_combined.head()
```

```python
scaled_data_combined.groupby('loan_status').count()['EDUCATION']
```

**SMOTE - Sythetic Minority Over-Sampling Technique** 

**Data Balancing**

```python
target = scaled_data_combined['loan_status']
features = scaled_data_combined.drop('loan_status', axis=1)
features.head()
```

```python
pip install imbalanced-learn
```

```python
from  imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
```

```python
balanced_features, balanced_target = smote.fit_resample(features,target)
```

```python
balanced_features.shape
```

```python
31671
```

```python
scaled_data_combined.groupby('loan_status').size()
```

```python
balanced_target_df = pd.DataFrame({'target':balanced_target})
balanced_target_df.groupby('target').size()
```

**Model Training**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(balanced_features, balanced_target, test_size=0.20,random_state=42)
```

**Building the models**

1. **Logistic regresion**

```python
logit = LogisticRegression()
```

```python
logit.fit(x_train,y_train)
```

```python
logit.score(x_train,y_train)
```

```python
logit_prediction = logit.predict(x_test)
```

```python
print(classification_report(y_test, logit_prediction))
```

```python
print(logit.coef_[0])
```

```python
# Lets create another Dataframe
features_imp_logit = pd.DataFrame({'features': balanced_features.columns, 'logit_imp':logit.coef_[0]})
features_imp_logit.sort_values(by='logit_imp', ascending=False)
```

2. **Random Forest**

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train, y_train)
```

```python
rf.score(x_train, y_train)
```

```python
rf_prediction = (rf.predict(x_test))
rf_prediction
```

```python
print(classification_report(y_test, rf_prediction))
```

```python
rf.feature_importances_
```

```python
# Df for RF
features_imp_rf = pd.DataFrame({'features':balanced_features.columns, 'rf_imp':rf.feature_importances_})
features_imp_rf.sort_values(by='rf_imp', ascending=False)
```

**XGBoost Model**

```python
xgb_model = XGBClassifier(tree_method = 'exact')
```

```python
# Train the model
# model.fit(x_train.values.ravel())
xgb_model.fit(x_train,y_train.values.ravel())
```

```python
xgb_model.score(x_train,y_train.values.ravel())
```

```python
# Making the predictions
xgb_prediction = xgb_model.predict(x_test)
xgb_prediction
```

```python
print(classification_report(y_test,xgb_prediction))
```

```python
features_imp_rf = pd.DataFrame({'features':balanced_features.columns, 'rf_imp':rf.feature_importances_})
features_imp_xgb = pd.DataFrame({'features':balanced_features.columns, 'xgb_imp':xgb_model.feature_importances_})
features_imp_xgb.sort_values(by='xgb_imp',ascending=False)
```

```python
features_imp = pd.concat([features_imp_logit,features_imp_rf,features_imp_xgb], axis=1)
features_imp
```
